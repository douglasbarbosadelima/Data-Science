```
import tensorflow as tf
import math
import numpy as np
import random
tf.__version__
```

Agora vamos criar sequências de 3 inteiros, começando com um a 15 deltas. O target será o quarto inteiro

```
X=[]
y=[]
random.seed(42)
for i in range(100):
  ini=random.randint(1,15)
  seq=[]
  for j in range(4):
    seq.append([ini+j])
  X.append(seq)
  y.append([ini+4])
X=np.array(X)
y=np.array(y)

X[:3],y[:3]
```

Modelo 1 : RNN com memória no Output

```
tf.random.set_seed(51)
model1=tf.keras.Sequential()
model1.add(tf.keras.layers.SimpleRNN(1,input_shape=(None,1),activation=lambda x:x))
```

```
model1.compile(loss='mean_squared_error', optimizer='adam')
```

```
model1.summary()
```

```
model1.fit(X, y, epochs=300)
```

```
w=model1.get_weights()
w
```

```
model1.predict([[[1.0],[2.0],[3.0],[4.0]]])
```

Modelo 2 : LSTM com ativação identidade

```
tf.random.set_seed(51)
model2=tf.keras.Sequential()
model2.add(tf.keras.layers.LSTM(1,input_shape=(None,1),activation=lambda x:x))
```

```
model2.compile(loss='mean_squared_error', optimizer='adam')
```

```
model2.fit(X, y, epochs=300)
```

```
w=model2.get_weights()
w
#ordem dos pesos input, recorrência e bias
#em cada vetor i,f,c,o
```

```
ww,wu,wb=w
ww,wu,wb
```

```
model2.predict([[[1.0],[2.0],[3.0],[4.0]]])
```

Modelo 3 : ANN com um neurônio e ativação identidade

```
#mudança para compatibilizar o input na ANN (o perceptron recebe os 4 inputs no mesmo t)
X3=X.reshape((100,1,4))
X3[:2]
```

```
tf.random.set_seed(51)
model3=tf.keras.Sequential()
model3.add(tf.keras.layers.Dense(1,input_shape=(1,4),activation=lambda x:x))
```

```
model3.compile(loss='mean_squared_error', optimizer='adam')
```

```
model3.fit(X3, y, epochs=300)
```

```
model3.get_weights()
```

```
model3.predict([[[1.0,2.0,3.0,4.0]]])
```
