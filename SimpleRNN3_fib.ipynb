# Simple RNN_3

```
from keras.models import Sequential
from keras.layers import SimpleRNN,Dense
from keras import optimizers
import numpy as np  
import random
from tensorflow import set_random_seed
```

```
set_random_seed(42)
np.random.seed(42)
random.seed(42)
X_train10=[]
Y_train10=[]
for i in range(500):
  i=random.randint(0,50)
  X_train10.append([[i],[i+1],[i+2],[i+3]])
  Y_train10.append([2 * i + 5])
X_train10=np.array(X_train10)
Y_train10=np.array(Y_train10)


```

```
# create and fit the SimpleRNN model
model = Sequential()
model.add(SimpleRNN(units=4,activation=None,input_shape=(None, 
    1)))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='rmsprop')
model.fit(X_train10, Y_train10, epochs=200,batch_size=10)
```

```
model.summary()
```

```
model.predict([[[[0], [1], [1], [2], [3], [5], [8]]]])
```
