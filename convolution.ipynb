# Convolution

```
import matplotlib.pyplot as plt
import numpy as np
img=plt.imread('http://www.linkededucation.com.br/ds2ipt/piano.png')
from scipy import signal
```

```
plt.imshow(img,cmap='gray')
```

```
img=img[:,:,0]
```

```

my_filter=np.array([[1, -1, 1],
                   [1, 2, 1],
                   [1, -1, 1]])


imgc =signal.convolve2d(img, my_filter, boundary='symm', mode='same')
```

```
plt.imshow(imgc,cmap='gray')
```

```
img[:3, :3]
```

```
(img[:3, :3] * my_filter).sum()
```
