```
import tensorflow as tf
from sklearn.feature_extraction.text import TfidfVectorizer,CountVectorizer
import pandas as pd
import numpy as np
import random
```

```
df=pd.read_csv('yelp_short_10k.csv')
df.head()
```

```
df.info()
```

```
df['stars'].value_counts()
```

```
df['classe'] = 1
```

```
ind = df[df['stars'] < 4].index
df.loc[ind, 'classe'] = 0

df['classe'].value_counts()
```

```
indd = random.sample(list(df[df['classe'] == 1].index), (len(df[df['classe'] == 1]) - len(df[df['classe'] == 0])))
```

```
df.drop(index=indd, inplace=True)
```

```
df['classe'].value_counts()
```

```
import nltk
nltk.download('stopwords')
stopw = nltk.corpus.stopwords.words('english')
```

```
vect = TfidfVectorizer(stop_words=stopw, max_features=12000, ngram_range=(1, 2))
vect
```

```
vect.fit(df['text'])
```

```
vect.vocabulary_
```

```
repx = vect.transform(df['text'])
repx = repx.toarray()

repx[:2]
```

```
len(vect.vocabulary_)
```

```
X = repx
y = np.array(df['classe'])

X, y
```

```
model = tf.keras.Sequential([
  tf.keras.layers.Dense(300, activation='relu', input_shape=(len(vect.vocabulary_),), kernel_regularizer=tf.keras.regularizers.l2(0.01)),
  tf.keras.layers.Dense(200, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.01)),
  tf.keras.layers.Dense(1, activation='sigmoid')
])
```

```
model.compile(optimzer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```
hist = model.fit(X, y, validation_split=0.2,epochs=15)
```

```
model.predict(vect.transform(['Very good. Amazing place']).toarray()) # A good phrase
model.predict(vect.transform(['This place is boring, terrible.']).toarray()) # A bad phrase
```
