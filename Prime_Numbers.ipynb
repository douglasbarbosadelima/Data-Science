```python
from keras.models import Sequential
from keras.layers import Dense
import numpy as np

# Next prime number

# Define the model architecture
model = Sequential([
    Dense(64, activation='relu', input_shape=(1,1)),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])
# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

LIMIT = 1000

def genPrimes (limit):
  primeList = [2]
  for i in range(3, limit):
    if not (i in primeList) and len(list(filter(lambda x: i%x == 0, primeList))) == 0:
      primeList.append(i)
  return [[1]] + list(map(lambda x: [x], primeList))

def genTest (primes, limit):
  primesAux = list(map(lambda x: x[0], primes))
  data = []
  for i in range(1, limit):
    data.append(1 if i in primesAux else 0)
  return list(map(lambda x: [x], data))

primeListTest = genPrimes(LIMIT*2)

X_train = np.array(list(map(lambda x: [x], np.arange(1, 1000, dtype=int))))
Y_train = np.array(genTest(primeListTest, LIMIT))

X_test = np.array(list(map(lambda x: [x], np.arange(1000, 2000, dtype=int))))
Y_test = np.array(genTest(primeListTest, LIMIT*2))[LIMIT:]

model.fit(X_train, Y_train, epochs=300, batch_size=32)
loss, accuracy = model.evaluate(X_test[:999], Y_test)

print('loss: ', loss, 'accuracy: ', accuracy)
```
