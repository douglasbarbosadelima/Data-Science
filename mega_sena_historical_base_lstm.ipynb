```python
from keras.models import Sequential
from keras.layers import LSTM, Dense
import pandas as pd
import numpy as np

!ls -la # Access to Machine's CLI.

sena_file = pd.read_excel('./sample_data/mega_sena_historical_base.xlsx')

sena_file = sena_file.iloc[::-1] # Reverse the dataset
sena_file.head()

model = Sequential()
model.add(LSTM(10, input_shape=(6, 1)))
model.add(Dense(6))

model.compile(loss='mse', optimizer='adam')

X_train = np.array(sena_file.iloc[:1500:2])
Y_train = np.array(sena_file.iloc[1:1500:2])

X_test = np.array(sena_file[1500:2590].iloc[::2])
Y_test = np.array(sena_file[1500:2590].iloc[1::2])

print('X_train len: ', len(X_train))
print('Y_train len: ', len(Y_train))
print('X_test len: ', len(X_test))
print('Y_test len: ', len(Y_test))

model.fit(X_train, Y_train, epochs=1500, batch_size=32, validation_data=(X_test, Y_test))
```

```python
predictions = model.predict([[15, 17, 28, 34, 35, 51]])
predictions = map(lambda x: abs(x), predictions)
list(predictions)
```
