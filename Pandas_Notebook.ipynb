```
carsFile = pd.read_csv('https://raw.githubusercontent.com/douglasbarbosadelima/Data-Science/master/cars-uci-linked.csv', squeeze= True)
carsFile.describe()
```

```
carsFile.dropna(inplace=True) # Drop missing columns :)
carsFile.head()
```

```
carsFile.info()
```

```
# Working with mean and standart

mediaPesos = carsFile['weight'].mean()
desvioPadrao = carsFile['weight'].std()
carsFile['weight'] = (carsFile['weight'] - mediaPesos) / desvioPadrao

carsFile['weight'].mean(), carsFile['weight'].std()
```

```
# Working with mean and standart

mediaPesos = carsFile['horsepower'].mean()
desvioPadrao = carsFile['horsepower'].std()
carsFile['horsepower'] = (carsFile['horsepower'] - mediaPesos) / desvioPadrao

carsFile['horsepower'].mean(), carsFile['horsepower'].std()
```

```
df = pd.DataFrame()
df['pot'] = carsFile['horsepower']
df['peso'] = carsFile['weight']
X = np.array(df)
Y = np.array(carsFile['mpg'])
X[:10]
```

```
plt.plot(X[:, 0], Y, 'bo')
plt.show()
```

```
ye = []
theta = [30, -7]
for i in range(len(X)):
  ye.append(theta[0] + theta[1] * X[i, 0])
ye = np.array(ye)

plt.plot(X[:, 1], Y, 'ro')
plt.plot(X[:, 1], ye, 'go')
plt.show()
```
